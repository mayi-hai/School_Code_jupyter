# python3计算TF-IDF值

```python
#导入书籍
import pandas as pd
# 有的环境配置下read_table出错，因此改用read_csv
raw = pd.read_csv("金庸-射雕英雄传txt精校版.txt",
                  names = ['txt'], sep ='aaa', encoding ="utf-8" ,engine='python')
# 章节判断用变量预处理
def m_head(tmpstr):
    return tmpstr[:1]
def m_mid(tmpstr):
    return tmpstr.find("回 ")
raw['head'] = raw.txt.apply(m_head)
raw['mid'] = raw.txt.apply(m_mid)
raw['len'] = raw.txt.apply(len)
# 章节判断
chapnum = 0
for i in range(len(raw)):
    if raw['head'][i] == "第" and raw['mid'][i] > 0 and raw['len'][i] < 30 :
        chapnum += 1
    if chapnum >= 40 and raw['txt'][i] == "附录一：成吉思汗家族" :
        chapnum = 0
    raw.loc[i, 'chap'] = chapnum  
# 删除临时变量
del raw['head']
del raw['mid']
del raw['len']
raw.head(0)
rawgrp = raw.groupby('chap')
chapter = rawgrp.sum()##.agg(sum) # 只有字符串列的情况下，sum函数自动转为合并字符串
chapter = chapter[chapter.index != 0]
#第一段内容
aim = chapter.txt[1]
#全文内容
article = ''
for i in range(1,41):
    article += chapter.txt[i];
```

```python
#去除停用词和分词
import jieba
import pandas as pd
stoplist = list(pd.read_csv('停用词.txt',
                    names = ['w'], sep = 'aaa', encoding = 'utf-8',engine='python').w)
aimParticiple=[ w for w in jieba.cut(aim) if w not in stoplist and len(w) > 1 ]
# def m_cut(article):
articleParticiple=[ w for w in jieba.cut(article) if w not in stoplist and len(w) > 1 ]
```

```python
#总词频统计
word_freq={}
for i in articleParticiple:
    if i not in word_freq:
        word_freq[i]=1
    else:
        word_freq[i]+=1
```

```python
#第一段词频统计
aim_freq={}
for i in aimParticiple:
    if i not in aim_freq:
        aim_freq[i]=1
    else:
        aim_freq[i]+=1
aim_freq
```

```python
import math
#计算每个词的IDF值
word_idf={}  #存储每个词的IDF值

for i in word_freq:
    k=0
    idf=0
    for j in range(1,41):
        if chapter.txt[j].find(str(i))>0:
            k=k+1
            idf=math.log(40/(k+1))
            word_idf[i]=idf
```

```python
word_idf
```

```python
#计算每个词的TF值
word_tf={}  #存储每个词的tf值
for i in aim_freq.keys():
        word_tf[i]=aim_freq[i]/len(aimParticiple)
```

# 用sklearn计算TF-IDF的值

```python
import pandas as pd
stoplist = list(pd.read_csv('停用词.txt',
                    names = ['w'], sep = 'aaa', encoding = 'utf-8',engine='python').w)
import jieba
def m_cut(aim):
    return [ w for w in jieba.cut(aim) if w not in stoplist and len(w) > 1 ]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer

txtlist = [ " ".join(m_cut(w)) for w in chapter.txt.iloc[:5]] 

vectorizer = CountVectorizer() 
X = vectorizer.fit_transform(txtlist) # 将文本中的词语转换为词频矩阵  

transformer = TfidfTransformer()  
tfidf = transformer.fit_transform(X)  #基于词频矩阵X计算TF-IDF值  
tfidf
```

```python
tfidf.toarray() # 转换为数组
```

```python
tfidf.todense() # 转换为矩阵
```
