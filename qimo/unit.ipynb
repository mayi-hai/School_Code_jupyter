```python
import pandas as pd
# 载入语料
raw = pd.read_csv("金庸-射雕英雄传txt精校版.txt",
                  names = ['txt'], sep ='aaa', encoding ="utf-8" ,engine='python')

# 章节判断用变量预处理
def m_head(tmpstr):
    return tmpstr[:1]

def m_mid(tmpstr):
    return tmpstr.find("回 ")

raw['head'] = raw.txt.apply(m_head)
raw['mid'] = raw.txt.apply(m_mid)
raw['len'] = raw.txt.apply(len)
# 章节判断
chapnum = 0
for i in range(len(raw)):
    if raw['head'][i] == "第" and raw['mid'][i] > 0 and raw['len'][i] < 30 :
        chapnum += 1
    if chapnum >= 40 and raw['txt'][i] == "附录一：成吉思汗家族" :
        chapnum = 0
    raw.loc[i, 'chap'] = chapnum
    
# 删除临时变量
del raw['head']
del raw['mid']
del raw['len']
# 提取出三四章内容
chap3=list(raw[raw.chap == 3].txt)
chap4=list(raw[raw.chap == 4].txt)

# chap3 = raw[raw['chap'] == 3].copy()
# chap3.reset_index(drop=True, inplace=True)
# chap3['paraidx'] = chap3.index
# del chap3['chap']
# del chap3['paraidx']

# chap4 = raw[raw['chap'] == 4].copy()
# chap4.reset_index(drop=True, inplace=True)
# chap4['paraidx'] = chap4.index
# del chap4['chap']
# del chap4['paraidx']


# 对三四章内容提取关键词
import jieba
import jieba.analyse
chap3_g=[]
chap4_g=[]
for w in chap3:
    chap3_g.append(jieba.analyse.extract_tags(w,20))
for w in chap4:
    chap4_g.append(jieba.analyse.extract_tags(w,20))

#选取训练集和测试集
text_chap3=chap3_g[120:150]
train_chap3=chap3_g[10:110]
text_chap4=chap4_g[150:180]
train_chap4=chap4_g[1:101]
```

```python
# 求PYA3,PYA4
#第三,四章关键词词频统计
a3=[]
for i in range(len(train_chap3)):    
    for j in train_chap3[i]:
        a3.append(j)
a4=[]
for i in range(len(train_chap4)):    
    for j in train_chap4[i]:
        a4.append(j)
b3={}
for i in a3:
    if i in b3:
        b3[i] += 1
    else:
        b3[i] = 1
b4={}
for i in a4:
    if i in b4:
        b4[i] += 1
    else:
        b4[i] = 1
c3=list(set(b3))
c4=list(set(b4))
PYA3={}
for i in c3:
    if i in b3:
        PYA3[i]=b3[i]/len(a3)
PYA4={}
for i in c4:
    if i in b4:
        PYA4[i]=b4[i]/len(a4)
```

```python
# 求PYA3,PYA4
#第三,四章关键词词频统计
a3=[]
for i in range(len(train_chap3)):    
    for j in train_chap3[i]:
        a3.append(j)
a4=[]
for i in range(len(train_chap4)):    
    for j in train_chap4[i]:
        a4.append(j)
b3={}
for i in a3:
    if i in b3:
        b3[i] += 1
    else:
        b3[i] = 1
b4={}
for i in a4:
    if i in b4:
        b4[i] += 1
    else:
        b4[i] = 1
c3=list(set(b3))
c4=list(set(b4))
PYA3={}
for i in c3:
    if i in b3:
        PYA3[i]=b3[i]/len(a3)
PYA4={}
for i in c4:
    if i in b4:
        PYA4[i]=b4[i]/len(a4)
```
