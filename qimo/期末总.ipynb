```python
import pandas as pd
# 载入语料
raw = pd.read_csv("金庸-射雕英雄传txt精校版.txt",
                  names = ['txt'], sep ='aaa', encoding ="utf-8" ,engine='python')

# 章节判断用变量预处理
def m_head(tmpstr):
    return tmpstr[:1]

def m_mid(tmpstr):
    return tmpstr.find("回 ")

raw['head'] = raw.txt.apply(m_head)
raw['mid'] = raw.txt.apply(m_mid)
raw['len'] = raw.txt.apply(len)
# 章节判断
chapnum = 0
for i in range(len(raw)):
    if raw['head'][i] == "第" and raw['mid'][i] > 0 and raw['len'][i] < 30 :
        chapnum += 1
    if chapnum >= 40 and raw['txt'][i] == "附录一：成吉思汗家族" :
        chapnum = 0
    raw.loc[i, 'chap'] = chapnum
    
# 删除临时变量
del raw['head']
del raw['mid']
del raw['len']
# 提取出三四章内容
chap3=list(raw[raw.chap == 3].txt)
chap4=list(raw[raw.chap == 4].txt)

# chap3 = raw[raw['chap'] == 3].copy()
# chap3.reset_index(drop=True, inplace=True)
# chap3['paraidx'] = chap3.index
# del chap3['chap']
# del chap3['paraidx']

# chap4 = raw[raw['chap'] == 4].copy()
# chap4.reset_index(drop=True, inplace=True)
# chap4['paraidx'] = chap4.index
# del chap4['chap']
# del chap4['paraidx']


# 对三四章内容提取关键词
import jieba
import jieba.analyse
chap3_g=[]
chap4_g=[]
for w in chap3:
    chap3_g.append(jieba.analyse.extract_tags(w,20))
for w in chap4:
    chap4_g.append(jieba.analyse.extract_tags(w,20))

#选取训练集和测试集
text_chap3=chap3_g[120:150]
train_chap3=chap3_g[10:110]
text_chap4=chap4_g[150:180]
train_chap4=chap4_g[1:101]
```

```python
# 求PYA3,PYA4
#第三,四章关键词词频统计
a3=[]
for i in range(len(train_chap3)):    
    for j in train_chap3[i]:
        a3.append(j)
a4=[]
for i in range(len(train_chap4)):    
    for j in train_chap4[i]:
        a4.append(j)
b3={}
for i in a3:
    if i in b3:
        b3[i] += 1
    else:
        b3[i] = 1
b4={}
for i in a4:
    if i in b4:
        b4[i] += 1
    else:
        b4[i] = 1
c3=list(set(b3))
c4=list(set(b4))
PYA3={}
for i in c3:
    if i in b3:
        PYA3[i]=b3[i]/len(a3)
PYA4={}
for i in c4:
    if i in b4:
        PYA4[i]=b4[i]/len(a4)
```

```python
# 对第三章训练集判断
PP3=[]
for j in range(30):
    P3 = {}
    for i in text_chap3[j]:
        if i in PYA3:
            P3[i]=PYA3[i]
        else:
            P3[i]=0.001
    PP3.append(P3)

Pl3=[]
for j in range(30):
    m=1
    for i in PP3[j]:
        m=m*PP3[j][i]
    Pl3.append(m)

PP4=[]
for j in range(30):
    P4 = {}
    for i in text_chap3[j]:
        if i in PYA4:
            P4[i]=PYA4[i]
        else:
            P4[i]=0.001
    PP4.append(P4)

Pl4=[]
for j in range(30):
    m=1
    for i in PP4[j]:
        m=m*PP4[j][i]
    Pl4.append(m)

import numpy as np
a=np.array(Pl3)
b=np.array(Pl4)
e=a>b
e
```

```python
# 对第四章训练集判断
BB3=[]
for j in range(30):
    B3 = {}
    for i in text_chap4[j]:
        if i in PYA3:
            B3[i]=PYA3[i]
        else:
            B3[i]=0.01
    BB3.append(B3)

Bl3=[]
for j in range(30):
    m=1
    for i in BB3[j]:
        m=m*BB3[j][i]
    Bl3.append(m)

BB4=[]
for j in range(30):
    B4 = {}
    for i in text_chap4[j]:
        if i in PYA4:
            B4[i]=PYA4[i]
        else:
            B4[i]=0.01
    BB4.append(B4)

Bl4=[]
for j in range(30):
    m=1
    for i in BB4[j]:
        m=m*BB4[j][i]
    Bl4.append(m)

import numpy as np
aa=np.array(Bl3)
ba=np.array(Bl4)
ea=aa>ba
ea
```
