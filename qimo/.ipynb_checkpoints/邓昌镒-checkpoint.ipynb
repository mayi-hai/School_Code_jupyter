# 第一问

```python
import pandas as pd
import jieba
import jieba.analyse
# 载入语料射雕英雄传
raw = pd.read_csv("附件2_金庸-射雕英雄传txt精校版.txt",
                  names = ['txt'], sep ='aaa', encoding ="utf-8" ,engine='python')
# 章节判断用变量预处理
def m_head(tmpstr):
    return tmpstr[:1]
def m_mid(tmpstr):
    return tmpstr.find("回 ")
raw['head'] = raw.txt.apply(m_head)
raw['mid'] = raw.txt.apply(m_mid)
raw['len'] = raw.txt.apply(len)
# 章节判断
chapnum = 0
for i in range(len(raw)):
    if raw['head'][i] == "第" and raw['mid'][i] > 0 and raw['len'][i] < 30 :
        chapnum += 1
    if chapnum >= 40 and raw['txt'][i] == "附录一：成吉思汗家族" :
        chapnum = 0
    raw.loc[i, 'chap'] = chapnum 
# 删除临时变量
del raw['head']
del raw['mid']
del raw['len']
# 提取出三四章内容
chap3=list(raw[raw.chap == 3].txt)
chap4=list(raw[raw.chap == 4].txt)
# 对三四章内容提取关键词
chap3_g=[]
chap4_g=[]
for w in chap3:
    chap3_g.append(jieba.analyse.extract_tags(w,20))
for w in chap4:
    chap4_g.append(jieba.analyse.extract_tags(w,20))
#选取训练集和测试集
test_chap3=chap3_g[150:190]
train_chap3=chap3_g[10:150]
test_chap4=chap4_g[150:190]
train_chap4=chap4_g[10:150]
# 进行训练集计算
# 求P(A|Y)=P(Y|A)*P(A)/P(Y)
a3=[]
for i in range(len(train_chap3)):    
    for j in train_chap3[i]:
        a3.append(j)
a4=[]
for i in range(len(train_chap4)):    
    for j in train_chap4[i]:
        a4.append(j)
b3={}
for i in a3:
    if i in b3:
        b3[i] += 1
    else:
        b3[i] = 1
b4={}
for i in a4:
    if i in b4:
        b4[i] += 1
    else:
        b4[i] = 1
c3=list(set(b3))
c4=list(set(b4))
PYA3={}
for i in c3:
    if i in b3:
        PYA3[i]=b3[i]/len(a3)
PYA4={}
for i in c4:
    if i in b4:
        PYA4[i]=b4[i]/len(a4)
# 对第三章测试集判断
PP3=[]
for j in range(30):
    P3 = {}
    for i in test_chap3[j]:
        if i in PYA3:
            P3[i]=PYA3[i]
        else:
            P3[i]=0.0007
    PP3.append(P3)
Pl3=[]
for j in range(30):
    m=1
    for i in PP3[j]:
        m=m*PP3[j][i]
    Pl3.append(m)
PP4=[]
for j in range(30):
    P4 = {}
    for i in test_chap3[j]:
        if i in PYA4:
            P4[i]=PYA4[i]
        else:
            P4[i]=0.0007
    PP4.append(P4)
Pl4=[]
for j in range(30):
    m=1
    for i in PP4[j]:
        m=m*PP4[j][i]
    Pl4.append(m)
# 对第四章测试集判断
BB3=[]
for j in range(30):
    B3 = {}
    for i in test_chap4[j]:
        if i in PYA3:
            B3[i]=PYA3[i]
        else:
            B3[i]=0.0007
    BB3.append(B3)
Bl3=[]
for j in range(30):
    m=1
    for i in BB3[j]:
        m=m*BB3[j][i]
    Bl3.append(m)
BB4=[]
for j in range(30):
    B4 = {}
    for i in test_chap4[j]:
        if i in PYA4:
            B4[i]=PYA4[i]
        else:
            B4[i]=0.0007
    BB4.append(B4)
Bl4=[]
for j in range(30):
    m=1
    for i in BB4[j]:
        m=m*BB4[j][i]
    Bl4.append(m)
# 进行判断，将测试集存入test列表中，将测试集的正确集存入true列表中
test=[]
true=[]
for i in range(len(Pl3)):
    if Pl3[i]>Pl4[i]:
        test.append('3')
    else:
        test.append("4")
    true.append('3')
for i in range(len(Bl3)):
    if Bl4[i]>Bl3[i]:
        test.append('4')
    else:
        test.append("3")
    true.append('4')
# 进行测验
from sklearn.metrics import classification_report
target_names = ['class 3', 'class 4',]
print("第一问\n",classification_report(true, test, target_names=target_names))
```

# 第二问

```python
import numpy as np
from tkinter import _flatten
# 载入语料射雕英雄传
raw1 = pd.read_csv("金庸-射雕英雄传txt精校版.txt",
                  names = ['txt'], sep ='aaa', encoding ="utf-8" ,engine='python')
# 载入语料天龙八部
raw2 = pd.read_csv("附件3_金庸-天龙八部txt精校版.txt",
                  names = ['txt'], sep ='aaa', encoding ="utf-8" ,engine='python')
#从天龙八部和射雕英雄传中各提取500段
chap1=raw1[1500:2000]
chap2=raw2[1500:2000]
# 将取出的段落转化为列表形式，sd为射雕英雄传，tl为天龙八部
para1_data = np.array(chap1)#np.ndarray()
para1=para1_data.tolist()#list
para2_data = np.array(chap2)#np.ndarray()
para2=para2_data.tolist()#list
sd = list(_flatten(para1))
tl = list(_flatten(para2))
# 提取关键字
import jieba
import jieba.analyse
sd_g=[]
tl_g=[]
for w in sd:
    sd_g.append(jieba.analyse.extract_tags(w,20))
for w in tl:
    tl_g.append(jieba.analyse.extract_tags(w,20))
#选取训练集350段和测试集150段
test_sd=sd_g[:150]
train_sd=sd_g[150:]
test_tl=tl_g[:150]
train_tl=tl_g[150:]
# 进行训练集计算
# 求P(A|Y)=P(Y|A)*P(A)/P(Y)
a1=[]
for i in range(len(train_sd)):    
    for j in train_sd[i]:
        a1.append(j)
a2=[]
for i in range(len(train_tl)):    
    for j in train_tl[i]:
        a2.append(j)
b1={}
for i in a1:
    if i in b1:
        b1[i] += 1
    else:
        b1[i] = 1
b2={}
for i in a2:
    if i in b2:
        b2[i] += 1
    else:
        b2[i] = 1
c1=list(set(b1))
c2=list(set(b2))
PYA1={}
for i in c1:
    if i in b1:
        PYA1[i]=b1[i]/len(a1)
PYA2={}
for i in c2:
    if i in b2:
        PYA2[i]=b2[i]/len(a2)
# 对射雕英雄传的测试集判断
PP1=[]
for j in range(150):
    P1 = {}
    for i in test_sd[j]:
        if i in PYA1:
            P1[i]=PYA1[i]
        else:
            P1[i]=0.0001
    PP1.append(P1)

Pl1=[]
for j in range(150):
    m=1
    for i in PP1[j]:
        m=m*PP1[j][i]
    Pl1.append(m)

PP2=[]
for j in range(150):
    P2 = {}
    for i in test_sd[j]:
        if i in PYA2:
            P2[i]=PYA2[i]
        else:
            P2[i]=0.0001
    PP2.append(P2)

Pl2=[]
for j in range(150):
    m=1
    for i in PP2[j]:
        m=m*PP2[j][i]
    Pl2.append(m)
# 对天龙八部的测试集判断
BB1=[]
for j in range(150):
    B1 = {}
    for i in test_tl[j]:
        if i in PYA1:
            B1[i]=PYA1[i]
        else:
            B1[i]=0.0001
    BB1.append(B1)

Bl1=[]
for j in range(150):
    m=1
    for i in BB1[j]:
        m=m*BB1[j][i]
    Bl1.append(m)

BB2=[]
for j in range(150):
    B2 = {}
    for i in test_tl[j]:
        if i in PYA2:
            B2[i]=PYA2[i]
        else:
            B2[i]=0.0001
    BB2.append(B2)

Bl2=[]
for j in range(150):
    m=1
    for i in BB2[j]:
        m=m*BB2[j][i]
    Bl2.append(m)
# 进行判断，将测试集存入test列表中，将测试集的正确集存入true列表中
test=[]
true=[]
for i in range(len(Pl1)):
    if Pl1[i]>Pl2[i]:
        test.append('1')
    else:
        test.append("2")
    true.append('1')
for i in range(len(Bl1)):
    if Bl2[i]>Bl1[i]:
        test.append('2')
    else:
        test.append("1")
    true.append('2')
from sklearn.metrics import classification_report
target_names = ['class 射雕', 'class 天龙',]
print("第二问\n",classification_report(true, test, target_names=target_names))
```
